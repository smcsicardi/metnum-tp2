### Comparacion de tiempos imagenes chicas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
test = pd.read_csv('csv/testVelocidadChicoRes.csv')
cantPersonas = test['cantPersonas']
promM = test['promM']
promMMonio = test['promMMonio']
df = pd.DataFrame()
df['Cantidad de personas'] = cantPersonas
df['Tiempos matriz M'] = promM
df['Tiempos matriz M-Moño'] = promMMonio

df.plot(x='Cantidad de personas')
plt.show()
```
